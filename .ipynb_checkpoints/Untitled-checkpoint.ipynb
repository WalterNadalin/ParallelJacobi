```python
from matplotlib.pyplot import subplots, show

a, b = (-0.9, 0.8)
c, d = (-0.2, 0.1)

fig, ax = subplots()
ax.hlines(y = 1, xmin = a, xmax = b, color = 'b', label = r'$[' + str(a) + ', ' + str(b) + ']$')
ax.hlines(y = 0, xmin = c, xmax = d, color = 'r', label = 'Seconda linea')
ax.set_xlim((-1, 1))
ax.set_ylim((-1, 2))
ax.legend(loc='center left', bbox_to_anchor = (1, 0.5))
ax.grid(linestyle = '--', axis = 'x')
ax.set_yticks(())
show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 100)
y = np.sin(x)

fig, ax = plt.subplots()

ax.plot(x, y)
ax.axhline(y=0.5, xmin=0.0, xmax=1.0, color='r')
ax.hlines(y=0.6, xmin=0.0, xmax=1.0, color='b')

plt.show()
```
